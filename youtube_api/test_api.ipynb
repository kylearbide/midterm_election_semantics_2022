```python
from googleapiclient.discovery import build
import datetime
from youtube_transcript_api import YouTubeTranscriptApi
import json
```

```python
with open('../config.json', 'r') as file:
    CONFIG = json.load(file)
api_key = CONFIG['youtube_API']
```

```python
youtube = build('youtube', 'v3', developerKey = api_key)
```

```python
today=datetime.date.today()
yesterday=today-datetime.timedelta(days=1)
daybefore=today-datetime.timedelta(days=2)
today_rfc3339=today.isoformat()+'T00:00:00+04:00' #+4:00 is needed to set to EST
yesterday_rfc3339=yesterday.isoformat()+'T00:00:00+04:00'
daybefore_rfc3339=daybefore.isoformat()+'T00:00:00+04:00'
```

```python
request = youtube.search().list(
        part="snippet",
        maxResults=2,
        publishedAfter=daybefore_rfc3339,
        publishedBefore=yesterday_rfc3339,
        q="ron johnson speech",
        type="video",
        videoCaption="videoCaptionUnspecified",
        videoDuration="long"
    )
```

```python
trump = request.execute()
```

```python
trump['items'][0]
```

```python
trumpID = trump['items'][1]['id']['videoId']
```

```python
request_trump = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=trumpID
        )
```

```python
response_trump = request_trump.execute()
```

```python
YouTubeTranscriptApi.get_transcript(trumpID)
```
